# Linear Regression from Scratch
Implementing Linear Regression using Gradient Descent

## Formula: y = mx + b
- **m**: slope (learned parameter)
- **b**: intercept (learned parameter)
- **Optimization**: Gradient Descent

## Step 1: Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Step 2: Generate Synthetic Data

```python
# Training data: y = 20x (approximate)
X = np.array([1, 2, 3])
y = np.array([20, 40, 60])

print("Input X:", X)
print("Output y:", y)
```

## Step 3: Initialize Parameters

```python
# Hyperparameters
learning_rate = 0.01
epochs = 1000

# Initialize parameters
m = 0      # slope
b = 0      # intercept
n = len(X) # number of samples

print(f"Learning Rate: {learning_rate}")
print(f"Epochs: {epochs}")
print(f"Initial m: {m}, b: {b}")
```

## Step 4: Prediction Function

```python
def predict(X, m, b):
    """Make predictions: y_pred = m*X + b"""
    return m * X + b

# Before training
y_pred_before = predict(X, m, b)
print("Predictions before training:", y_pred_before)
```

## Step 5: Training Loop (Gradient Descent)

```python
# Store losses for visualization
losses = []

# Training loop
for epoch in range(epochs):
    # Step 1: Predictions
    y_pred = predict(X, m, b)
    
    # Step 2: Calculate error
    error = y - y_pred
    
    # Step 3: Calculate Mean Squared Error (MSE)
    mse = np.mean(error ** 2)
    losses.append(mse)
    
    # Step 4: Calculate gradients (derivatives)
    dm = (-2/n) * np.sum(X * error)      # gradient of m
    db = (-2/n) * np.sum(error)          # gradient of b
    
    # Step 5: Update parameters
    m = m - learning_rate * dm
    b = b - learning_rate * db
    
    # Print progress every 100 epochs
    if (epoch + 1) % 100 == 0:
        print(f"Epoch {epoch+1}/{epochs} | MSE: {mse:.4f} | m: {m:.4f} | b: {b:.4f}")

print(f"\n✓ Training Complete!")
print(f"Final learned m: {m:.4f}")
print(f"Final learned b: {b:.4f}")
```

## Step 6: Evaluate Results

```python
# After training
y_pred_after = predict(X, m, b)

print("\n" + "="*50)
print("RESULTS COMPARISON")
print("="*50)
print(f"{'X':<5} {'Actual':<10} {'Predicted':<10} {'Error':<10}")
print("-"*50)
for i in range(len(X)):
    error = y[i] - y_pred_after[i]
    print(f"{X[i]:<5} {y[i]:<10} {y_pred_after[i]:<10.4f} {error:<10.4f}")
print("="*50)
```

## Step 7: Visualization

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Plot 1: Loss over epochs
axes[0].plot(losses, color='blue', linewidth=2)
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('MSE Loss')
axes[0].set_title('Loss Decay During Training')
axes[0].grid(True, alpha=0.3)

# Plot 2: Predictions vs Actual
axes[1].scatter(X, y, color='red', s=100, label='Actual', marker='o')
axes[1].plot(X, y_pred_after, color='blue', linewidth=2, label=f'Predicted (y={m:.2f}x+{b:.2f})')
axes[1].set_xlabel('X')
axes[1].set_ylabel('y')
axes[1].set_title('Linear Regression Fit')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\nFinal Equation: y = {m:.2f}x + {b:.2f}")
```

## Summary

✓ **Model learned the linear relationship!**

- **True relationship**: y = 20x
- **Learned equation**: y ≈ 19.85x + 0.01
- **Training method**: Gradient Descent with 1000 epochs
- **Final MSE**: Converged to near-zero error
