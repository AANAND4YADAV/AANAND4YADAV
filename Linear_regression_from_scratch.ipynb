```python
import numpy as np
class LinearRegression:
    def __init__(self,learning_rate=0.01,epochs=1000):
        self.lr=learning_rate
        self.epochs=epochs
        # storing hyparameters
        self.m=0
        self.b=0  #initializing m and b to 0

    def fit(self,X,y):
        n=len(X)
        for _ in range(self.epochs):
            y_pred=self.m*X+self.b
            dm=(-2/n)*np.sum(X*(y-y_pred))
            db=(-2/n)*np.sum(y-y_pred)
            self.m-=self.lr*dm
            self.b-=self.lr*db
            # loss is calculated but not used after the loop, so it's fine

    def predict(self, X):
        return self.m*X+self.b

#testing the implementation
if __name__=="__main__":
    #generating synthetic data
    X=np.array([1,2,3])
    y=np.array([20,40,60])  #y=2x
    model=LinearRegression(learning_rate=0.01,epochs=1000)
    #predictions=model.fit(X,y)
    model.m=10 # These initializations might be overwritten by fit if m and b are re-initialized inside fit
    model.b=0
    print("Before training :",model.predict(X))
    model.fit(X,y)
    print("After training :", model.predict(X))
    print(" learned m :", model.m)
```
